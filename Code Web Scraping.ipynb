```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.geeksforgeeks.org/data-preprocessing-machine-learning-python/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
data = soup.find_all('div', class_ = 'highlight')
```

```python
code = data[6].text.strip()
code
```

```python
code_update = code.split('\n')
code_update
```

```python
for code in code_update:
    print(code)
```
